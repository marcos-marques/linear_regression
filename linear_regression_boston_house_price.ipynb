### Linear Regression with statsmodels

```python
import statsmodels.api as sm
from sklearn import datasets
import numpy as np
import pandas as pd
```

Para criação do modelo de Regressão Linear utilizamos o conjunto de dados disponível no sklearn

```python
# Carregando o conjunto de dados do preço de casas em Boston
data = datasets.load_boston()

# Criando um DataFrame com os dados de conjunto de dados do Sklearn.
# O Sklearn já define quais colunas são as variáveis independentes, então podemos passar "feature_names" como colunas de features
df_boston = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df_boston.head()
```

```python
# Da mesma forma podemos obter nossa coluna alvo
target = pd.DataFrame(data.target, columns=['MEDV'])
```

```python
# Para um primeiro teste estamos utilizando a variável "RM" que diz respeito ao número médio de quartos. Essa será nossa variável independee
X = df_boston['RM']
y = target['MEDV']
```

```python
# Em seguida fazemos os fit dos dados para gerar nosso modelo
model = sm.OLS(y, X).fit()
```

```python
# Uma vez finalizado nosso modelo, podemos fazer uma previsão
predictions = model.predict(X)
```

```python
print(model.summary())
```

### Entendendo a Tabela:

- **Dep. Variable:** Variável dependente
- **Model:** Ordinary Least Squares (método do tipo mínimos quadrados linear) para estimar parâmetros desconhecidos em um modelo de regresssão linear.
- **Df Residuals:** O número de valores no cálculo final de uma estatística que são livres para variar
- **Coef:** O valor $3.6534$ significa que conforme a variável **RM** aumenta em 1, o valor previsto de **MDEV** aumenta $3.6534$
- **R ao quadrado:** a porcentagem de variância que o modelo explica. 
- **std er:** É o desvio padrão da distribuição amostral de uma estatística, normalmente da média.
